```python
import pandas as pd
```

```python
data_g = pd.read_csv("G_loss.csv").transpose()
data_d = pd.read_csv('D_loss.csv').transpose()
```

```python
data_g = data_g.stack().reset_index()
data_d = data_d.stack().reset_index()
```

```python
data_g[0].shape
```

```python
import numpy as np

gen_loss_sum = np.array(data_g[0])
dis_loss_sum = np.array(data_d[0])
print(gen_loss_sum.shape)
```

```python
import matplotlib.pyplot as plt

# plt.figure((15,15))
x = np.arange(9162)
plt.figure(figsize=(16,10))
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.plot(x, gen_loss_sum)
plt.plot(x, dis_loss_sum)
plt.savefig('loss.png')
plt.show()
```

```python
gen_loss_sum_sort = sorted(gen_loss_sum, reverse = True)
```

```python
plt.plot(x, gen_loss_sum_sort)
plt.show()
```
